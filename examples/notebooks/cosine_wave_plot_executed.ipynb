## Cosine Wave Plot

This notebook generates and displays a cosine wave using NumPy and Matplotlib. It creates x values spanning multiple periods and plots the cosine of those values to illustrate sinusoidal behavior. You can modify the frequency, amplitude, or range of x to explore different waveforms.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate values from 0 to 20 (10 periods)
x = np.linspace(0, 20 * np.pi, 500)
# Calculate cosine values
y = np.cos(x)

# Create the plot and capture the figure handle per guardrails
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, y, label='cos(x)', color='blue')
ax.set_title('Cosine Wave - 10 Periods')
ax.set_xlabel('x (radians)')
ax.set_ylabel('cos(x)')
ax.axhline(0, color='black', lw=0.5, ls='--')
ax.axvline(0, color='black', lw=0.5, ls='--')
ax.legend()
ax.grid()

# Save the figure to the specified path and close the figure
fig.savefig('/Users/indro/Projects/AgenTUI/examples/images/cosine_wave_10_periods.png', dpi=200)
plt.show()
plt.close(fig)
```

```python
import matplotlib.patches as patches

# Plot an ellipse and save it alongside the cosine image
fig, ax = plt.subplots(figsize=(6,6))
# Create an ellipse centered at (0,0) with width 4 and height 2, rotated 30 degrees
ellipse = patches.Ellipse((0, 0), width=4, height=2, angle=30, edgecolor='red', facecolor='none', lw=2)
ax.add_patch(ellipse)
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.set_aspect('equal')
ax.set_title('Ellipse Example')
ax.grid(True)

# Save and close per guardrails
ellipse_path = '/Users/indro/Projects/AgenTUI/examples/images/ellipse_example.png'
fig.savefig(ellipse_path, dpi=200)
plt.show()
plt.close(fig)
print(f"Saved ellipse image to: {ellipse_path}")
```
