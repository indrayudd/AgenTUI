## Cosine Wave Plot

This notebook generates and displays a cosine wave using NumPy and Matplotlib. It creates x values spanning multiple periods and plots the cosine of those values to illustrate sinusoidal behavior. You can modify the frequency, amplitude, or range of x to explore different waveforms.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate values from 0 to 20 (10 periods)
x = np.linspace(0, 20 * np.pi, 500)
# Calculate cosine values
y = np.cos(x)

# Create the plot and capture the figure handle per guardrails
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, y, label='cos(x)', color='blue')
ax.set_title('Cosine Wave - 10 Periods')
ax.set_xlabel('x (radians)')
ax.set_ylabel('cos(x)')
ax.axhline(0, color='black', lw=0.5, ls='--')
ax.axvline(0, color='black', lw=0.5, ls='--')
ax.legend()
ax.grid()

# Save the figure to the specified path and close the figure
fig.savefig('/Users/indro/Projects/AgenTUI/examples/images/cosine_wave_10_periods.png', dpi=200)
plt.show()
plt.close(fig)
```

```python
# Plot a hyperbola and save it alongside the other figures
import numpy as np
import matplotlib.pyplot as plt
import os

out_dir = '/Users/indro/Projects/AgenTUI/examples/images'
if not os.path.exists(out_dir):
    os.makedirs(out_dir, exist_ok=True)

# Hyperbola: x^2/a^2 - y^2/b^2 = 1 -> y = sqrt(x^2/a^2 - 1) * b
a = 1.0
b = 0.5
x_pos = np.linspace(a * 1.01, 5, 400)
x_neg = np.linspace(-5, -a * 1.01, 400)

y_pos = b * np.sqrt((x_pos**2) / (a**2) - 1)
y_neg = b * np.sqrt((x_neg**2) / (a**2) - 1)

fig, ax = plt.subplots(figsize=(6,6))
ax.plot(x_pos, y_pos, color='C0', label='branch +')
ax.plot(x_pos, -y_pos, color='C0', label='branch -')
ax.plot(x_neg, y_neg, color='C1')
ax.plot(x_neg, -y_neg, color='C1')

ax.set_aspect('equal', adjustable='box')
ax.set_xlim(-5,5)
ax.set_ylim(-3,3)
ax.set_title('Hyperbola: $x^2/a^2 - y^2/b^2 = 1$')
ax.legend()

hyperbola_path = os.path.join(out_dir, 'hyperbola_plot.png')
fig.savefig(hyperbola_path, dpi=200)
plt.show()
plt.close(fig)

print(f"Saved hyperbola image to: {hyperbola_path}")
```
