## Sine plot

```python
# Sine plot
import numpy as np
import matplotlib.pyplot as plt

# Create data
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x)

# Plot and save
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Sine wave')
fig.savefig('/Users/indro/Projects/AgenTUI/examples/images/sine.png', dpi=200)
plt.close(fig)

# Display confirmation
print('Saved /Users/indro/Projects/AgenTUI/examples/images/sine.png')
```
