## Cosine Wave Plot
This notebook plots a cosine wave with annotations for maxima and minima.

```python
import numpy as npimport matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 4 * np.pi, 500)y = np.cos(x)maxima_x = 0maxima_y = 1minima_x = np.piminima_y = -1fig, ax = plt.subplots(figsize=(8, 4))ax.plot(x, y, label='cos(x)')ax.scatter([maxima_x, minima_x], [maxima_y, minima_y], color='red')ax.annotate('max', (maxima_x, maxima_y), textcoords='offset points', xytext=(0, 10), ha='center')ax.annotate('min', (minima_x, minima_y), textcoords='offset points', xytext=(0, -15), ha='center')ax.set_title('Cosine Wave')ax.set_xlabel('x')ax.set_ylabel('cos(x)')ax.grid(True)ax.legend()fig.savefig('artifacts/cosine_wave.png', dpi=200)plt.show()plt.close(fig)
```
