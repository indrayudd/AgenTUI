## Exponential Curve Plotting

```
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 5, 100)
y = np.exp(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='y = e^x', color='blue')
plt.title('Exponential Curve')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid()
plt.legend()

# Save the plot
plt.savefig('/Users/indro/Projects/AgenTUI/examples/images/exponential_curve.png')
plt.show()
```
