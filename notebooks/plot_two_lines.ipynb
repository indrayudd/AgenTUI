## Plot two lines

```
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(-10, 10, 100)
y1 = 5*x + 2
y2 = 7*x - 6

# Plot
fig, ax = plt.subplots()
ax.plot(x, y1, label='y=5x+2')
ax.plot(x, y2, label='y=7x-6')
ax.legend()
ax.set_title('Plot of two lines')
ax.set_xlabel('x')
ax.set_ylabel('y')

# Save figure
fig.savefig('/Users/indro/Projects/AgenTUI/examples/images/two_lines_plot.png', dpi=200)
plt.close(fig)
```

```
# Autogenerated summary cell
m1, b1 = 5, 2
m2, b2 = 7, -6
# compute intersection
x0 = (b2 - b1) / (m1 - m2)
y0 = m1 * x0 + b1

def eq_str(m, b):
    return f"y={m}x{b:+}"

print(f"The lines {eq_str(m1,b1)} and {eq_str(m2,b2)} intersect at ({x0:.4g}, {y0:.4g}). "
      f"A perpendicular to the x-axis at x={x0:.4g} was drawn.")
```
